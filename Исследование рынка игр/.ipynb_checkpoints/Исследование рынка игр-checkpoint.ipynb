```python
no_scores = {}
platf = list(df['platform'].unique())
for plat in platf:
    no_scores[plat] = df[(df['user_score'].isna()) & (df['platform'] == plat)]['user_score'].shape[0]
no_scores = pd.DataFrame([no_scores]).T.reset_index()
no_scores.columns = ['platform', 'no_score']
```

```python
no_scores.sort_values(by='no_score', ascending=False).plot(kind='bar', x='platform', y='no_score',
                                                          figsize=(10,6), grid=True, 
                                                           title='–ö–æ–ª–∏—á–µ—Å—Ç–≤–æ –∏–≥—Ä, –Ω–µ –∏–º–µ—é—â–µ–π –æ—Ü–µ–Ω–∫–∏ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—è, –ø–æ –ø–ª–∞—Ç—Ñ–æ—Ä–º–∞–º');
```

–ë–æ–ª—å—à–∏–Ω—Å—Ç–≤–æ –∏–≥—Ä –±–µ–∑ –æ—Ü–µ–Ω–∫–∏ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π - –Ω–∞ –ø–ª–∞—Ç—Ñ–æ—Ä–º–µ PS (PS2, PSP, PS3)

```python
no_scores = {}
platf = list(df['platform'].unique())
for plat in platf:
    no_scores[plat] = df[(df['critic_score'].isna()) & (df['platform'] == plat)]['critic_score'].shape[0]
no_scores = pd.DataFrame([no_scores]).T.reset_index()
no_scores.columns = ['platform', 'no_score']
no_scores.sort_values(by='no_score', ascending=False).plot(kind='bar', x='platform', y='no_score',
                                                          figsize=(10,6), grid=True, 
                                                           title='–ö–æ–ª–∏—á–µ—Å—Ç–≤–æ –∏–≥—Ä, –Ω–µ –∏–º–µ—é—â–µ–π –æ—Ü–µ–Ω–∫–∏ –∫—Ä–∏—Ç–∏–∫–æ–≤, –ø–æ –ø–ª–∞—Ç—Ñ–æ—Ä–º–∞–º',
                                                          ylabel='–ö–æ–ª–∏—á–µ—Å—Ç–≤–æ');
```

–ö—Ä–∏—Ç–∏–∫–∏ —Å–≤–æ–∏–º –≤–Ω–∏–º–∞–Ω–∏–µ –≤ –±–æ–ª—å—à–∏–Ω—Å—Ç–≤–≤–µ —Å–ª—É—á–∞–µ–≤ –æ–±–¥–µ–ª—è—é—Ç —Ç–∞–∫–∏–µ –ø–ª–∞—Ç—Ñ–æ—Ä–º—ã –∫–∞–∫, DS, PS, PS2 PSP –∏ WII

```python
sales_cols = ['na_sales', 'eu_sales', 'jp_sales', 'other_sales']
no_score_sales_share = df.query('user_score != user_score or critic_score != critic_score')[sales_cols].sum() / df[sales_cols].sum()
print(
    "–î–æ–ª–∏ –ø—Ä–æ–¥–∞–∂ –∏–≥—Ä –±–µ–∑ –æ—Ü–µ–Ω–æ–∫ –≤ —Ä–∞–º–∫–∞—Ö —Ä–µ–≥–∏–æ–Ω–æ–≤:",
    ",".join([f' {name.split("_")[0]}: {share:.2%}' for name, share in no_score_sales_share.iteritems()]))
```

–í —Ü–µ–ª–æ–º –≤—ã—Ö–æ–¥–∏—Ç —á—Ç–æ –ø–æ–∫–æ–ª–µ–Ω–∏–µ –ø–ª–∞—Ç—Ñ–æ—Ä–º —Å–æ—Å—Ç–∞–≤–ª—è–µ—Ç –æ–∫–æ–ª–æ 10 –ª–µ—Ç, —Å –º–æ–º–µ–Ω—Ç–∞ –∑–∞—Ä–æ–∂–¥–µ–Ω–∏—è –∏ –¥–æ –≤—ã—Ö–æ–¥–∞ –Ω–∞ "–ø–µ–Ω—Å–∏—é".

–í –¥–∞–Ω–Ω–æ–º –¥–∞—Ç–∞—Å–µ—Ç–µ, –Ω–∞ –¥–æ—Å—Ç–ø–Ω—ã–π –≤—Ä–µ–º–µ–Ω–Ω–æ–π –ø–µ—Ä–∏–æ–¥, –ø—Ä–∏—Å—É—Ç—Å–≤—É–µ—Ç —Ç—Ä–µ–Ω–¥ –∫ —É–º–µ–Ω—å—à–µ–Ω–∏—é –ø—Ä–æ–¥–∞–∂, —ç—Ç–æ —Ö–∞—Ä–∞–∫—Ç–µ—Ä–Ω–æ –¥–ª—è –∫–∞–∂–¥–æ–π –ø–ª–∞—Ç—Ñ–æ—Ä–º—ã –∏–∑ —Ç–æ–ø 5

```python
year_pivot = df.pivot_table(index='platform', values='year_of_release', aggfunc=['min', 'max'])
year_pivot.columns = ['min_year', 'max_year']
year_pivot.reset_index(inplace=True)
year_pivot['platf_life'] = year_pivot['max_year'] - year_pivot['min_year']
year_pivot
```

```python
print(f'–ú–µ–¥–∏–∞–Ω–Ω–æ–µ –≤—Ä–µ–º—è –∂–∏–∑–Ω–∏ –ø–ª–∞—Ç—Ñ–æ—Ä–º—ã —Å—Ä–µ–¥–∏ –≤—Å–µ—Ö –ø–ª–∞—Ç—Ñ–æ—Ä–º: {year_pivot["platf_life"].median()}')
print(f'–°—Ä–µ–¥–Ω–µ–µ –≤—Ä–µ–º—è –∂–∏–∑–Ω–∏ –ø–ª–∞—Ç—Ñ–æ—Ä–º—ã —Å—Ä–µ–¥–∏ –≤—Å–µ—Ö –ø–ª–∞—Ç—Ñ–æ—Ä–º: {round(year_pivot["platf_life"].mean(),2)}')
```

```python
top_5 = df[df['platform'].isin(platform['platform'].values)]
year_pivot = top_5.pivot_table(index='platform', values='year_of_release', aggfunc=['min', 'max'])
year_pivot.columns = ['min_year', 'max_year']
year_pivot.reset_index(inplace=True)
year_pivot['platf_life'] = year_pivot['max_year'] - year_pivot['min_year']
year_pivot
```

```python
print(f'–ú–µ–¥–∏–∞–Ω–Ω–æ–µ –≤—Ä–µ–º—è –∂–∏–∑–Ω–∏ –ø–ª–∞—Ç—Ñ–æ—Ä–º—ã —Å—Ä–µ–¥–∏ —Ç–æ–ø 5 –ø–ª–∞—Ç—Ñ–æ—Ä–º: {year_pivot["platf_life"].median()}')
print(f'–°—Ä–µ–¥–Ω–µ–µ –≤—Ä–µ–º—è –∂–∏–∑–Ω–∏ –ø–ª–∞—Ç—Ñ–æ—Ä–º—ã —Å—Ä–µ–¥–∏ —Ç–æ–ø 5 –ø–ª–∞—Ç—Ñ–æ—Ä–º: {round(year_pivot["platf_life"].mean(),2)}')
```

–ò—Å–∫–ª—é—á–∏–º DS, —É –Ω–µ–µ –∫–∞–∫ —Ç–æ –Ω–µ–æ–±—ã—á–Ω–æ –º–Ω–æ–≥–æ –ª–µ—Ç —Å—Ä–æ–∫ –∂–∏–∑–Ω–∏ –∏–¥–µ—Ç

```python
year_pivot = year_pivot[year_pivot['platform'] != 'DS']
year_pivot
```

```python
stats = {}
stats['platform'] = []
stats['mean_sales'] = []
stats['median_sales'] = []
stats['min_sales'] = []
stats['max_sales'] = []
stats['std_sales'] = []
for platf in top_platforms:
    draw = top_platforms_data[top_platforms_data['platform'] == platf]
    stats['platform'].append(platf)
    stats['mean_sales'].append(round(draw['all_sales'].mean(),2))
    stats['median_sales'].append(round(draw['all_sales'].median(),2))
    stats['min_sales'].append(round(draw['all_sales'].min(),2))
    stats['max_sales'].append(round(draw['all_sales'].max(),2))
    stats['std_sales'].append(round(draw['all_sales'].var(),2))
stats = pd.DataFrame(stats).set_index('platform').sort_values(by='median_sales', ascending=False)
stats
```

```python
genre_stats = {}
genre_stats['genre'] = []
genre_stats['mean_sales'] = []
genre_stats['median_sales'] = []
genre_stats['min_sales'] = []
genre_stats['max_sales'] = []
genre_stats['std_sales'] = []
for genre in list(df['genre'].unique()):
    draw = data[data['genre'] == genre]
    genre_stats['genre'].append(genre)
    genre_stats['mean_sales'].append(round(draw['all_sales'].mean(),2))
    genre_stats['median_sales'].append(round(draw['all_sales'].median(),2))
    genre_stats['min_sales'].append(round(draw['all_sales'].min(),2))
    genre_stats['max_sales'].append(round(draw['all_sales'].max(),2))
    genre_stats['std_sales'].append(round(draw['all_sales'].var(),2))
genre_stats = pd.DataFrame(genre_stats).set_index('genre').sort_values(by='median_sales', ascending=False)
genre_stats
```

- H0 –°—Ä–µ–¥–Ω–∏–µ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—å—Å–∫–∏–µ —Ä–µ–π—Ç–∏–Ω–≥–∏ –Ω–∞ –ø–ª–∞—Ç—Ñ–æ—Ä–º–µ XboxOne –∏ PC –æ–¥–∏–Ω–∞–∫–æ–≤—ã–µ
- H1 –°—Ä–µ–¥–Ω–∏–µ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—å—Å–∫–∏–µ —Ä–µ–π—Ç–∏–Ω–≥–∏ –Ω–∞ –ø–ª–∞—Ç—Ñ–æ—Ä–º–µ XboxOne –∏ PC —Ä–∞–∑–Ω—ã–µ

```python
xbox_one = df[(df['platform'] == 'XOne') & (df['user_score'] != 'tbd')]
xbox_one['user_score'] = xbox_one['user_score'].apply(float)
xbox_one['user_score'].mean()
```

```python
pc = df[(df['platform'] == 'PC') & (df['user_score'] != 'tbd')]
pc['user_score'] = pc['user_score'].apply(float)
pc['user_score'].mean()
```

```python
alpha = 0.05

results = st.ttest_ind(
    xbox_one['user_score'].dropna(), 
    pc['user_score'].dropna())

print('p-–∑–Ω–∞—á–µ–Ω–∏–µ:', results.pvalue)

if results.pvalue < alpha:
    print("–û—Ç–≤–µ—Ä–≥–∞–µ–º –Ω—É–ª–µ–≤—É—é –≥–∏–ø–æ—Ç–µ–∑—É")
else:
    print("–ù–µ –ø–æ–ª—É—á–∏–ª–æ—Å—å –æ—Ç–≤–µ—Ä–≥–Ω—É—Ç—å –Ω—É–ª–µ–≤—É—é –≥–∏–ø–æ—Ç–µ–∑—É")
```
