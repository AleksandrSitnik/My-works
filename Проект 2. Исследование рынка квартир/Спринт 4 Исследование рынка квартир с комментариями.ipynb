Не особо видно зависимости, хотя по моему опыту должна быть, в оценочной деятельности это называется фактор масштаба )

```python
def area_group(row):
    if row['total_area'] <= 20:
        return '< 20'
    if 20 < row['total_area'] <= 40 :
        return '21-40'
    if 41 < row['total_area'] <= 60 :
        return '41-60'
    if 61 < row['total_area'] <= 90 :
        return '61-90'
    if 91 < row['total_area'] <= 120 :
        return '91-120'
    if 121 < row['total_area'] <= 160 :
        return '121-160'
    return '160+'

data['area_type'] = data.apply(area_group, axis=1)
```
